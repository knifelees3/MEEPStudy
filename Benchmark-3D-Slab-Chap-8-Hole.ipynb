```python
from meep import mpb
import numpy as np
import meep as mp
import matplotlib.pyplot as plt
#from mayavi import mlab

resolution = 15  # pixels/a

a = 1         # units of um
r = 0.3*a         # units of um
h = 0.6*a        # units of um
w = 1*a        # units of um

r = r/a          # units of "a"
h = h/a          # units of "a"
w = w/a          # units of "a"


Si = mp.Medium(epsilon=12)
size = mp.Vector3(a,a,9*a)
geometry_lattice = mp.Lattice(size=size,
                              basis1=mp.Vector3(np.sqrt(3)/2,0.5),
                              basis2=mp.Vector3(np.sqrt(3)/2,-0.5),)
default_material = Si
# 
geometry = [ mp.Block(material=Si,size=(a,a,h),
                      center=mp.Vector3(),
                      e1=mp.cartesian_to_lattice(mp.Vector3(1,0,0),geometry_lattice),
                      e2=mp.cartesian_to_lattice(mp.Vector3(0,1,0),geometry_lattice),
                      e3=mp.cartesian_to_lattice(mp.Vector3(0,0,1),geometry_lattice),),
             mp.Cylinder(center=mp.Vector3(),radius=r, height=h, material=mp.air)]


num_k = 10
k_points = mp.interpolate(num_k, [mp.Vector3(0,0,), mp.Vector3(0.5,0),
                                  mp.Vector3(1/3,-1/3), mp.Vector3(0,0,)])

num_bands = 6

ms = mpb.ModeSolver(geometry_lattice=geometry_lattice,
                    geometry=geometry,
                    k_points=k_points,
                    resolution=resolution,
                    num_bands=num_bands)
ms.run_te()

# freq = ms.all_freqs
# plt.plot(np.linspace(0,1,freq.shape[0]),freq[:,0:])

# plt.show()
```
