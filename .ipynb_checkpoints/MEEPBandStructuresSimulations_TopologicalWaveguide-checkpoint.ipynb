有了前面的准备，我们现在开始用MEEP计算更加高端的拓扑光子晶体波导结构:

主要参考文献为：
Chen, Y. et al. Topologically Protected Valley-Dependent Quantum Photonic Circuits. Phys. Rev. Lett. 126, 230503 (2021).
其结构参数如下：

>Our nanophotonic structures are fabricated on SOI wafers with 220-nm-thick silicon layers by electron-beam lithography (for more details, see Supplemental Material [31]). The valley-dependent photonic topological structures comprise two kinds of hexagonal-profile air holes of different side lengths arranged in a honeycomb lattice, which break the spatial-inversion symmetry of the system [shown in Figs. 1(a) and 1(b)]. First, we study the bulk topology of the transverse-electric (TE)-like band [shown in Fig. 1(c)] by using the MIT Photonic Bands (MPB) package [32] to calculate the band structures of the unit cell of the lattice [dotted line in Figs. 1(a) and 1(b)]. The side lengths of the two hexagonal air holes are s1 = 87 nm and s2 = 127 nm. The lattice constant is a =470 nm. This special design opens a TE-like polarization band gap between 1520 and 1600 nm.



六角结构，
$s_{1}=87 nm, s_{2}=127nm,a=470 nm$,根据文章结果，我们主要关注TE模式即可

能带周期方向以及结构设置如上图，介质折射率为2.48

```python
import math
import meep as mp
from meep import mpb
import numpy as np

num_bands = 8

# 定义好扫描的k波矢
k_points = [mp.Vector3(),          # Gamma
            mp.Vector3(0.5),       # K
            mp.Vector3(3.0/8,np.sqrt(3)/8 ),  # M
            mp.Vector3()]          # Gamma

# 对k波矢进行插值
k_points = mp.interpolate(20, k_points)

#  units of nm
s1=87
s2=127
a=470
# units of a
s1=s1/a
s2=s1/a


# 基本结构和材料性质
# 折射率设置
nslab = 2.36
slab = mp.Medium(index=nslab)

nair = 1
air = mp.Medium(index=nair)


# 设置一个六角结构
# A hexagonal prism defined by six vertices centered on the origin
# of material crystalline silicon (from the materials library)
vertices1 = [mp.Vector3(-1*s1,0),
            mp.Vector3(-0.5*s1,math.sqrt(3)/2*s1),
            mp.Vector3(0.5*s1,math.sqrt(3)/2*s1),
            mp.Vector3(1*s1,0),
            mp.Vector3(0.5*s1,-math.sqrt(3)/2*s1),
            mp.Vector3(-0.5*s1,-math.sqrt(3)/2*s1)]

vertices2 = [mp.Vector3(-1*s2,0),
            mp.Vector3(-0.5*s2,math.sqrt(3)/2*s2),
            mp.Vector3(0.5*s2,math.sqrt(3)/2*s2),
            mp.Vector3(1*s2,0),
            mp.Vector3(0.5*s2,-math.sqrt(3)/2*s2),
            mp.Vector3(-0.5*s2,-math.sqrt(3)/2*s2)]

geometry = [mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(0,0), material=air),
            
           mp.Prism(vertices2, height=mp.inf, center=mp.Vector3(np.sqrt(3)/2,1/2), material=air),
           mp.Prism(vertices2, height=mp.inf, center=mp.Vector3(-np.sqrt(3)/2,1/2), material=air),
           mp.Prism(vertices2, height=mp.inf, center=mp.Vector3(0,-1), material=air)]

# geometry = [mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(0,0), material=air),
#             mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(1,0), material=air),
#             mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(-1,0), material=air),
           
#            mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(0.5,np.sqrt(3)/2), material=air),
#            mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(-0.5,np.sqrt(3)/2), material=air), 
#            mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(-1.5,-np.sqrt(3)/2), material=air),
            
#            mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(0.5,-np.sqrt(3)/2), material=air),
#            mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(-0.5,-np.sqrt(3)/2), material=air),
#            mp.Prism(vertices1, height=mp.inf, center=mp.Vector3(-1.5,-np.sqrt(3)/2), material=air),
            
#            mp.Prism(vertices2, height=mp.inf, center=mp.Vector3(np.sqrt(3)/2,-1/2), material=air),
#            mp.Prism(vertices2, height=mp.inf, center=mp.Vector3(-np.sqrt(3)/2,-1/2), material=air),
            
#            mp.Prism(vertices2, height=mp.inf, center=mp.Vector3(-1,1), material=air),
#            mp.Prism(vertices2, height=mp.inf, center=mp.Vector3(0,1), material=air),
#            mp.Prism(vertices2, height=mp.inf, center=mp.Vector3(1,1), material=air)]

geometry_lattice = mp.Lattice(
    size=mp.Vector3(1, 1),
    basis1=mp.Vector3(0.5, np.sqrt(3) / 2),
    basis2=mp.Vector3(1,0),
)
```

```python
resolution = 32
# mode solver设置
ms = mpb.ModeSolver(num_bands=num_bands,
                    k_points=k_points,
                    geometry=geometry,
                    geometry_lattice=geometry_lattice,
                    default_material=slab,
                    resolution=resolution)
```

```python
ms.run_te()
te_freqs = ms.all_freqs
te_gaps = ms.gap_list
ms.run_tm()
tm_freqs = ms.all_freqs
tm_gaps = ms.gap_list
```

```python


import matplotlib.pyplot as plt
numk,tmp=np.shape(te_freqs)
klist=range(numk)

fig,ax=plt.subplots()
x = range(len(te_freqs))

for l in range(num_bands):
    plt.plot(te_freqs[:,l],'r-')
    plt.plot(tm_freqs[:,l],'b-')
plt.ylim([0,1.2])
plt.ylabel('$\omega a/(2\pi c)$')

# Plot gaps
for gap in te_gaps:
    if gap[0] > 1:
        ax.fill_between(x, gap[1], gap[2], color='blue', alpha=0.2)
for gap in tm_gaps:
    if gap[0] > 1:
        ax.fill_between(x, gap[1], gap[2], color='blue', alpha=0.2)
        

points_in_between = (len(te_freqs) - 4) / 3
tick_locs = [i*points_in_between+i for i in range(4)]
tick_labs = ['Γ', 'K', 'M', 'Γ']
ax.set_xticks(tick_locs)
ax.set_xticklabels(tick_labs, size=16)
```
