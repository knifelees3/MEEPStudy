计算六角晶格，采用最小的平行四边形结构单元。



其主要结构如上图所示，


绿色的原胞内，两个柱子的坐标为

$$
x_{c1}=\frac{1}{2}R, y_{c1}=\frac{\sqrt(3)}{2}R
$$

$$
x_{c2}=\frac{\sqrt(3)a-R}{2}, y_{c2}=\sqrt(3)\frac{\sqrt(3)a-R}{2}
$$

可以采用绿色的结构，也可以采用红色的原胞。和直角的结果不同，对于三角结构，其倒格矢与原基矢有旋转，我们需要用到倒格矢与原基矢量的关系：
$$
\left\{\begin{matrix}  \boldsymbol{b}_1=\frac{2\pi  \boldsymbol{a}_2\times \boldsymbol{a}_3}{\boldsymbol{a}_1\cdot (\boldsymbol{a}_2\times \boldsymbol{a}_3)}  \\  \boldsymbol{b}_2=\frac{2\pi  \boldsymbol{a}_3\times \boldsymbol{a}_1}{\boldsymbol{a}_1\cdot (\boldsymbol{a}_2\times \boldsymbol{a}_3)}  \\  \boldsymbol{b}_3=\frac{2\pi  \boldsymbol{a}_1\times \boldsymbol{a}_2}{\boldsymbol{a}_1\cdot (\boldsymbol{a}_2\times \boldsymbol{a}_3)}  \end{matrix}\right.\\
$$
其中分母的物理意义代表着原胞的体积。计算图中结构时，需要首先确定最小原胞，可以有多种选取方式，按照图中的选取方法，最小原胞是一个平行四边形，原胞的两个基矢为

$$
a_{1}=(\frac{\sqrt{3}}{2},\frac{1}{2})a,a_{2}=(0,1)a
$$


按照图中的定义，可以计算出倒格子空间的矢量为

$$
b_{1}=(\frac{2}{\sqrt{3}},0)\frac{2\pi}{a},b_{2}=(-\frac{1}{\sqrt{3}},1)\frac{2\pi}{a}
$$

我们计算能带，需要改变波矢来求频率本征值。**在MEEP中，我们一旦手动指定了倒格矢的方向，接下里的能带的`kpoint`都是以该基矢展开的，这与COMSOL是不同的。因此我们需要给出特定点用倒格矢表示的坐标。**

按照图中的定义，$\Gamma,\kappa,M$三个点为的坐标为，

其他点的坐标分别为

$$
(0,0),\quad (\frac{1}{\sqrt{3}},0), \quad (\frac{1}{\sqrt{3}},\frac{1}{2\sqrt{3}})
$$

到此，所有的准备工作就做完了，接下来可以进行代码实现了。

## MPB计算

```python
import math
import meep as mp
from meep import mpb
import numpy as np
import matplotlib.pyplot as plt

num_bands = 7

# 定义好扫描的k波矢
k_points = [mp.Vector3(1/np.sqrt(3),0),   # K
            mp.Vector3(),          # Gamma
            mp.Vector3(1/np.sqrt(3),1/np.sqrt(3)/2),   # M
            mp.Vector3(1/np.sqrt(3),0)]   # K

# 对k波矢进行插值
k_points = mp.interpolate(40, k_points)

#  units of nm
a=np.sqrt(3)  # Lattice constants
a0=a/np.sqrt(3)

eta=3.125
R=a/eta # radius of hexagonal shape

d=2*a/9 
r=d/2 # radius of rod

# 基本结构和材料性质
# 折射率设置
nslab = np.sqrt(11.7)
slab = mp.Medium(index=nslab)

nair = 1
air = mp.Medium(index=nair)


geometry_lattice = mp.Lattice(
    size=mp.Vector3(a0, a0),
    basis1=mp.Vector3( np.sqrt(3) / 2,0.5),
    basis2=mp.Vector3(0, 1),
)

xc1=1/2*R
yc1=np.sqrt(3)/2*R
xc2=(np.sqrt(3)*a0-R)/2
yc2=np.sqrt(3)*(np.sqrt(3)*a0-R)/2

geometry = [mp.Cylinder(r,center=mp.cartesian_to_lattice(mp.Vector3(xc1,yc1),geometry_lattice), material=slab),
            mp.Cylinder(r,center=mp.cartesian_to_lattice(mp.Vector3(xc2,yc2),geometry_lattice), material=slab),
            ]


resolution = 20

# mode solver设置
ms = mpb.ModeSolver(num_bands=num_bands,
                    k_points=k_points,
                    geometry=geometry,
                    geometry_lattice=geometry_lattice,
                    resolution=resolution)

ms.run_tm()
tm_freqs = ms.all_freqs
tm_gaps = ms.gap_list
```

# 导入COMSOL结果

```python
# 导入COMSOl数据
with open('untitled.txt', 'r') as file:
    data = []
    for line in file:
        # 假设每行有一个科学计数法的数值
        value = float(line.strip())  # 转换为浮动数值
        data.append(value)


dataCOMSOL=np.reshape(np.array(data),(33,7))
print(np.shape(dataCOMSOL))
```
